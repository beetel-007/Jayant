# Histogram


```python
plt.hist(data.Income)
```

```python
plt.hist(data.Income,bins=20) #hist-histogram
```

```python
plt.hist(data.Income,bins=20,color='green') #bins-to use more prescribe graph
```

```python
plt.figure(figsize=(15,7)) #to set size of the graph
plt.hist(data.Income,bins=20,color='Black',alpha=0.8) #alpha-to provide trasnparency to graph
plt.xlabel("income")
plt.ylabel("frequency(no of employees)")
plt.title("Histogram.income")
plt.show()
```

```python
from matplotlib import style #to add effect or details in the graph
```

```python
style.use("classic")
plt.figure(figsize=(15,7)) #to set size of the graph
plt.hist(data.Income,bins=20,color='green',alpha=0.8) #alpha-to provide trasnparency to graph
plt.xlabel("income")
plt.ylabel("frequency(no of employees)")
plt.title("Histogram.income")
plt.tight_layout()
#plt.savefig("histogram.pdf")
```

# Density plot


```python
data['Income'].plot(kind='density', ls="--") #ls-line style
```

```python
data['Income'].plot(kind='density')
plt.xlabel("income")
plt.ylabel("frequency")
plt.title("density")
plt.tight_layout
plt.figure(figsize=(15,7))
```

# Histogram with density plot



```python
ax=data['Income'].plot(kind='hist',color='lightblue')
ax2=data['Income'].plot(kind='density', secondary_y=True,color='blue')
ax.set_ylabel("frequency(no of employees)")
ax2.set_ylabel("Density")
ax.set_xlabel("Income")
```
